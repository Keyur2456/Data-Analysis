## Convert String Datetime to Date

```python
station_day['Date'] = pd.to_datetime(station_day.Date)
station_hour['Datetime'] = pd.to_datetime(station_hour.Datetime)
madrid_data['date'] = pd.to_datetime(madrid_data.date)
```

```python
df1 = station_day.groupby([station_day['Date'].dt.year.rename('year')])['AQI'].max()
print(df1.head)
plt.plot(df1)
plt.title("Max AQI based on year", fontsize = 15)
plt.xlabel("Year", fontsize = 12)
plt.ylabel("AQI", fontsize = 12)
plt.show()
```

```python
df1 = station_day.groupby([station_day['Date'].dt.year.rename('year')])['AQI'].mean()
print(df1.head)
plt.plot(df1)
plt.title("Average AQI based on year", fontsize = 15)
plt.xlabel("Year", fontsize = 12)
plt.ylabel("AQI", fontsize = 12)
plt.show()
```

```python
madrid_data_CO_YEARLY = madrid_data.groupby([madrid_data['date'].dt.year.rename('year')])['CO'].mean()
madrid_data_CO_YEARLY
```

```python
india_station_data_CO_YEARLY = station_day.groupby([station_day['Date'].dt.year.rename('year')])['CO'].mean()
india_station_data_CO_YEARLY
```

```python
plt.plot(india_station_data_CO_YEARLY, "o")
plt.plot(madrid_data_CO_YEARLY, "o")
plt.xlabel("Year")
plt.ylabel("Carbon Monoxide in mg / m3")
plt.title("CO level by year in both countries")
plt.legend(['India', 'Madrid'])
plt.show()
```
